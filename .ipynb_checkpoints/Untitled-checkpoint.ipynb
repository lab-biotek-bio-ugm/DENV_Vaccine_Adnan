```python
import requests, sys
import json
```

```python
requestURL = "https://www.ebi.ac.uk/proteins/api/coordinates?offset=0&size=100&gene=FGFR2"

r = requests.get(requestURL, headers={ "Accept" : "application/json"})

if not r.ok:
    r.raise_for_status()
    sys.exit()

responseBody = r.text
parsed_json = json.loads(responseBody)
```

```python
import bs4
from bs4 import BeautifulSoup

with open("VaxiJen - predicting protective antigens.html") as f:
    read_data = f.read()

soup = BeautifulSoup(read_data, 'html.parser')
```

```python
t = soup.find_all('font')
b = soup.find_all('b')
result = []
for i in range(int(len(t)/3)):
    try:
        y = 3 + (i*3)
        score = (b[y].contents)
        score = float(score[0].replace(' ',''))
        ID = t[y].contents
        ID = ID[0].replace('\n', '')
        seqx = t[y+1].contents
        seqfin = []
        for x in seqx:
            if type(x) == (bs4.element.NavigableString):
                seqfin.append(x)
        seq = "".join(seqfin)
        seq = str(seq.replace(' ',''))    
        data = [score, ID, seq]
        result.append(data)
    except IndexError as error:
        pass
result
```
