# Analyze VaxiJen Result

Q: VaxiJen digunakan untuk apa? Jelaskan input dan ouput dari Vaxijen!

A: analisis sekuens protein yang paling efisien sebagai antigen. Input 232 sekuens dalam .fas, output 1 sekuens dengan skor tertinggi. Threshold > 0.5

Reference Paper: https://doi.org/10.1186/1471-2105-8-4


```python
with open("VaxiJen - predicting protective antigens.html") as f:
    read_data = f.read()
soup = BeautifulSoup(read_data, 'html.parser')
t = soup.find_all('font')
b = soup.find_all('b')
result = []
for i in range(int(len(t)/3)):
    try:
        y = 3 + (i*3)
        score = (b[y].contents)
        score = float(score[0].replace(' ',''))
        ID = t[y].contents
        ID = ID[0].replace('\n', '')
        seqx = t[y+1].contents
        seqfin = []
        for x in seqx:
            if type(x) == (bs4.element.NavigableString):
                seqfin.append(x)
        seq = "".join(seqfin)
        seq = str(seq.replace(' ',''))    
        data = [score, ID, seq]
        result.append(data)
    except IndexError as error:
        pass
#result[0]
```

```python
df2 = pd.DataFrame(result, columns = ["Score", "ID", "Sequence"])
df2.sort_values(by=['Score'], ascending=False)
```

# Analyse NetCTL result
Q: NetCTL digunakan untuk apa? Jelaskan input dan output dari NetCTL!

A: prediksi epitop terbaik dari sekuens B4XPM1 (output vaxijen). Input: sekuens B4XPM1.fas. Output: combinatorial score terbaik dari masing-masing supertype (total 12) yang nantinya dipilih 5 terbaik. Threshold: 0.5, 0.89 dan 0.94 berturut-turut.

Reference Paper:https://doi.org/10.1186/1471-2105-8-424

```python
with open("NetCTL-1.2 Server - prediction results.html") as f2:
    read_data2 = f2.read()
soup2 = BeautifulSoup(read_data2, 'html.parser')
net = soup2.find_all('pre')
net = net[0].contents
net = str(net[0])
net = net.split("\n")
net_result = net[2:-8]
net_hit = []
for i in net_result:
    data = i.split()
    if len(data) > 15:
        final_data = [int(data[0]), data[2], data[4], float(data[6]), float(data[8]), float(data[10]), float(data[12]), float(data[14])]
        net_hit.append(final_data)
#net_hit
```

```python
df3 = pd.DataFrame(net_hit, columns = ["Res No", "ID", "sequence", "aff", "aff_rescale", "cle", "tap", "COMB"])
df3.sort_values(by=['COMB'], ascending=False)
```
